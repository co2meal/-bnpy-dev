# Demo-HashtagK9

## Goal
This notebook demonstrates using **bnpy** to fit a Gaussian mixture model via EM.

```python
import os
import numpy as np
import bnpy

from matplotlib import pylab
%pylab inline
```

## Data

We'll use the HashtagK9 dataset, which comes by default with **bnpy**.

Since this is a dataset with only a few thousand 2D observations, we can make a simple scatter plot.

```python
import HashtagK9
Data = HashtagK9.get_data(nObsTotal=20000)
pylab.plot(Data.X[:,0], Data.X[:,1], 'k.');
pylab.gcf().set_size_inches(8, 4)
pylab.axis('image'); 
pylab.xlim([-10, 10]); pylab.ylim([-5, 5]);
```

## Single Run with K=10

Here, we show how to train a Gaussian mixture with 10 components via EM.

```python
hmodel, RInfo = bnpy.run(Data, 'MixModel', 'Gauss', 'EM', 
                jobname='demo-K10', K=10, nLap=500, convergeThr=0.01, printEvery=25)
```

Each of the 10 learned clusters has an estimated mean, and an estimated covariance matrix.  

We can plot each cluster with elliptical contours.  See bnpy/viz/GaussViz.py for details.

```python
bnpy.viz.PlotComps.plotCompsForTask('HashtagK9/demo-K10/1/');
pylab.gcf().set_size_inches(8, 4);
pylab.axis('image'); 
pylab.xlim([-10, 10]); pylab.ylim([-5, 5]);
```

## Training models across a range of K values

We now explore how models of different complexity perform on this dataset.

Generally, as K increases we see *better evidence scores*, even as $K$ goes well beyond the number of ideal clusters (K=9). Above $K=9$ the evidence score improvements are very tiny, but visible.

```python
for K in [4, 6, 8, 10, 15, 20]:
    hmodel, RInfo = bnpy.run(Data, 'MixModel', 'Gauss', 'EM',
                             jobname='demo-K=%d' %(K), K=K, 
                             nLap=200, convergeThr=0.01, 
                             doWriteStdOut=False)
    print 'K=%3d | final ev % .5f' % (K, RInfo['evBound'])
```

## Plot: K vs final ELBO score

```python
bnpy.viz.PlotELBO.plotJobsThatMatchKeywords('HashtagK9/demo-K=*', xvar='K');
pylab.ylim([-2.5, -1.5]);
```

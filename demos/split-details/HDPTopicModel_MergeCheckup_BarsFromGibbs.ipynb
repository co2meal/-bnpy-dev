Define function to construct proposal given a single pair of topics to merge.

```python
def constructMergeModelAndLP(Data, curModel, curLP, kA, kB):
    ''' Create proposal variational parameters (global and local), given current ones.
    
    Returns
    -------
    propModel : bnpy model, with updated global params and K-1 comps
    propLP : dict of local params, with updated values and K-1 comps
    '''
    # Construct the proposed beta vector
    propBeta = np.delete(beta, kB)
    propBeta[kA] += beta[kB]
    # Set all allocation model parameters from this beta
    propModel = curModel.copy()
    propModel.allocModel.ClearCache()
    propModel.allocModel.K = K - 1
    propModel.allocModel.rho = bnpy.util.StickBreakUtil.beta2rho(propBeta, K-1)
    propModel.allocModel.omega = (Data.nDoc + curModel.allocModel.gamma) * np.ones(K-1)
    # Construct observed suff stats, and do merge
    obsSS = curModel.obsModel.get_global_suff_stats(Data, None, curLP)
    obsSS.mergeComps(kA, kB)
    propModel.obsModel.update_global_params(obsSS)
    
    # Now construct local parameters
    propResp = np.delete(LP['resp'], kB, axis=1)
    propResp[:, kA] += LP['resp'][:, kB]
    propLP = propModel.allocModel.initLPFromResp(Data, dict(resp=propResp), alphaEbeta=alpha*propBeta, alphaEbetaRem=alpha*betaRem)
    
    return propModel, propLP
```

Define function that can compute the ELBO for a particular configuration (global params in model object, local params in LP dictionary)

```python
from bnpy.allocmodel.topics.HDPTopicModel import calcELBO as calcLrest

def calcELBO(Data, model, LP, multiplier=1.0/100, todict=0):
    # Compute Ldata and all other terms of the ELBO, and piece it all together
    obsSS = model.obsModel.get_global_suff_stats(Data, None, LP)
    Lrest = calcLrest(Data=Data, LP=LP, rho=model.allocModel.rho, omega=model.allocModel.omega,
                      alpha=alpha, gamma=gamma, todict=todict)
    Ldata = model.obsModel.calc_evidence(None, obsSS, None, todict=todict)
    Lscale = model.obsModel.getDatasetScale(obsSS) * multiplier
    if todict:
        Ldict = dict(**Ldata)
        Ldict.update(Lrest)
        ukeys = list(set([key.split('_')[0] for key in Ldict.keys()]))
        for key in Ldict.keys():
            if key not in ukeys:
                del Ldict[key]
        Ltotal = 0
        for key in ukeys:
            Ldict[key] /= Lscale
            Ltotal += Ldict[key]
        Ldict['Ltotal'] = Ltotal
        return Ldict
    else:
        return (Lrest + Ldata) / Lscale
```

Define function for plotting change of each ELBO term

```python
def plotBarChartOfGainInELBOAfterMerge(propLdict, curLdict, ukeys, W=0.4, hpad=0.25):
    pylab.figure()
    xs = np.linspace(0, (1.0 + hpad*W) * len(ukeys), len(ukeys))
    barvals = np.asarray([propLdict[key]-curLdict[key] for key in ukeys])
    barHandles = pylab.bar(xs, barvals, width=W);
    for k, barH in enumerate(barHandles):
        if barvals[k] < 0:
            barH.set_color('r')
        else:
            barH.set_color('g')
            
        height = barH.get_height()
        pylab.text(barH.get_x()+barH.get_width()/2., 1.05*height, '% .2f'%(barvals[k]),
                ha='center', va='bottom')

    # Set Y lims
    B = 1.25 * np.max(np.abs(barvals))
    pylab.ylim([-B, B]);
    # Set X lims and ticks
    pylab.gca().set_xticks(xs+W/2);
    pylab.gca().set_xticklabels(ukeys);
    # Set size (needs to be so wide to read the xlabels
    pylab.gcf().set_size_inches((10,2));

    if propLdict['Ltotal'] > curLdict['Ltotal']:
        pylab.title('ACCEPTED !');
    else:
        pylab.title('rejected :(');
```

Define function that ranks all prospective partners for a comp, by gain in Ldata term.

```python
ukeys = ['Ltotal', 'Ldata', 'Lentropy', 'LcDtheta', 'Lalloc', 'Lslack']

def tryTopMergeProposalsByLdataForTopic(Data, hmodel, LP, kA, multiplier=1.0/100):
    curLdict = calcELBO(Data, hmodel, LP, todict=1)

    obsSS = hmodel.obsModel.get_global_suff_stats(Data, None, LP)
    mPairs_kA = [(np.minimum(kA, x), np.maximum(kA, x)) for x in range(SS.K) if x != kA]
    gainL = hmodel.obsModel.calcHardMergeGap_SpecificPairs(obsSS, mPairs_kA)
    scale = hmodel.obsModel.getDatasetScale(SS) * multiplier
    gainL /= scale
    
    bigtosmall = np.argsort(-1 * gainL)
    for rankk, k in enumerate(bigtosmall[:3]):
        print "Rank %d: pair %s" % (rankk, mPairs_kA[k])
        print "   predicted gain for Ldata: % .5f" % (gainL[k])

        k1, k2 = mPairs_kA[k]
        if k1 == kA:
            kpartner = k2
        else:
            kpartner = k1
        
        propModel, propLP = constructMergeModelAndLP(Data, hmodel, LP, k1, k2)
        propLdict = calcELBO(Data, propModel, propLP, todict=1)
        print "   actual gain for Ldata :   % .5f" % (propLdict['Ldata'] - curLdict['Ldata'])
        print "   actual gain for Ltotal:   % .5f" % (propLdict['Ltotal'] - curLdict['Ltotal'])
            
        # Plot the two topics we are merging
        figH = bnpy.viz.BarsViz.showTopicsAsSquareImages(
            SS.WordCounts[[kA, kpartner]], vmin=0, vmax=10, xlabels=['ktarget=%d' % (kA), 'kpartner=%d'%(kpartner)]);

        # Plot the ELBO gains
        plotBarChartOfGainInELBOAfterMerge(propLdict, curLdict, ukeys)
```

# Create current model from stored run of the Gibbs sampler stuck with some junk / redundant topics

This includes the local parameters (resp, theta) and the global parameters (rho, omega, likelihood params).

```python
hmodel, Info = bnpy.run(Data, 'HDPTopicModel', 'Mult', 'moVB', 
                  nLap=3, nBatch=1, 
                  alpha=alpha,
                  gamma=gamma,
                  initname='/data/liv/x-topics/aistats-final/BarsK10/aistats-alg=teh-K=50/1/',
                  jobname='tmpfromteh',
                  **LPkwargs);
```

```python
# Post process to remove empty slots
LP = hmodel.calc_local_params(Data, **LPkwargs)
SS = hmodel.get_global_suff_stats(Data, LP)
bigtosmall = np.argsort(-1 * SS.getCountVec())
SS.reorderComps(bigtosmall)
smallComps = np.flatnonzero(SS.getCountVec() < 10)
for k in reversed(smallComps):
    SS.removeComp(k)
hmodel.update_global_params(SS)
```

```python
K = hmodel.allocModel.K
betaRem = 0.01
beta = (1.0 - betaRem) / K * np.ones(K)
assert np.allclose(betaRem + beta.sum(), 1.0)
hmodel.allocModel.ClearCache()
hmodel.allocModel.rho = bnpy.util.StickBreakUtil.beta2rho(beta, K)
hmodel.allocModel.omega = (Data.nDoc + hmodel.allocModel.gamma) * np.ones(K)
```

```python
LP = hmodel.calc_local_params(Data, **LPkwargs)
SS = hmodel.get_global_suff_stats(Data, LP)
hmodel.obsModel.update_global_params(SS)
```

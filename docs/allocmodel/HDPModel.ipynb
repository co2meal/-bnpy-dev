# HDPModel

## GOAL

This is documentation for HDPModel, a **bnpy** allocation model.

This document only addresses *allocation*, which means we will only
describe notations and update equations that pertain to assigning data
items to clusters/components. Anything to do with the likelihood
funciton is covered in the appropriate **obsmodel** documentation.

## Notation

### Global Parameters

Let vector $\beta$ give the top-level probabilities of each topic (including a "remaining mass" term at index $K+1$)

$$
\beta = [\beta_1 ~ \beta_2 ~ \ldots \beta_K ~\beta_{K+1} ]
$$

Let vector $\pi_d$ give the topic probabilities within document $d$ (including a "remaining mass" term at index $K+1$)

$$
\pi_d = [\pi_{d1} ~ \pi_{d2} ~ \ldots \pi_{dK} ~ \pi_{d,K+1} ]
$$

### Local parameters


We let $z_{dv}$ denote the assignment of words of type $v$ in document $d$ to one of the $K$ active topics.

Here, we'll use *one-hot* notation, in which $z_{dv}$ is a **binary** vector of length $K$.

$$
z_{dv} = [ z_{dv1} ~ z_{dv2} \ldots z_{dvK} ]
$$

This *one-hot* vector (aka indicator vector) has two constraints

* each entry is binary: $z_{dvk} \in \{0, 1\}$
* the vector sums to one: $\sum_{k=1}^K z_{dvk} = 1$

These constraints force only one entry in $z_{dvk}$ to be "on", with the remaining entries all off.

## Local posterior

$$
p( z_{dv} \mid x, \pi_d, \phi_k ) \approx q(z_{dv} \mid r_{dv}) = \mbox{Discrete}( z_{dv} \mid r_{dv1}, r_{dv2}, \ldots r_{dvK} )
$$

The vector $r_{dv}$ must obey two constraints

* each entry is non-negative: $r_{dvk} \ge 0$
* sums-to-one: $\sum_{k=1}^K r_{dvk} = 1$

$$
p( \pi_d \mid z_d, \beta) \approx q( \pi_d \mid \theta_d ) = \mbox{Dirichlet}( \pi_d \mid \theta_{d1}, \ldots \theta_{dK}, \theta_{dK+1} )
$$

The vector $\theta_{d}$ must obey one constraint

-  each entry is non-negative: $\theta_{dk} \ge 0$


## Local posterior updates (variational E-step)

We need to alternate two updates

* $q(z_d)$ given $q(\pi_d)$, aka $r_d$ given $\theta_d$
* $q(\pi_d)$ given $q(z_d)$, aka $\theta_d$ given $r_d$


### Update to local assignments

First, we find the log of $r$ up to a normalization constant.

$\begin{align}
\log \tilde{r}_{dvk} &= \mbox{E}_{q(\pi), q(\phi)} \Big[ \log p( z_{dvk} = 1 \mid x_{dv}, \pi_d, \phi) \Big] 
\\ 
&= \mbox{E}_{\pi}[ \log p( z_{dvk}=1 \mid \pi_d) ] + \mbox{E}_{\phi}[ \log p( x_{dv} \mid z_{dvk}=1, \phi) ] + \mbox{const} 
\\ 
&= \mbox{E}[ \log \pi_{dk} ] + \mbox{E}[ \log \phi_{kv} ]
\end{align}$

Next, we exponentiate and normalize, to obtain an $r$ that satisfies its constraints.

$$
r_{dvk} = \frac{ \exp[ \log \tilde{r}_{dvk} ] }{\sum_{\ell=1}^K \exp[ \log \tilde{r}_{dv\ell} ] }
$$

### Update to document-level probabilities


### Coordinate-ascent algorithm

**while** not converged

* for each doc $d$:
* * $\mbox{expElog}[\pi_d] \gets \mbox{exp} ( \mbox{E}[ \log \pi_{dk} ] )$
* * $\mbox{sum}[\tilde{r}_{d}] \gets \mbox{MatrixVectorProduct}( \mbox{exp}[ L_{d} ], \mbox{expElog}[\pi_d] )$
* * $N_{d,\cdot} \gets \mbox{MatrixVectorProduct}( \mbox{exp}[ L_{d} ]^T, c_d / \tilde{r}_d )$
* $N_{d,\cdot} \gets N_{d, \cdot} \times \mbox{expElog}[\pi_d]$


$\begin{align}
N_{dk} 
&\triangleq \sum_{n=1}^{U_d} c_{dn} r_{dnk}
\\
&= \sum_{n=1}^{U_d} c_{dn} \frac{\tilde{r}_{dnk}}{\sum_{\ell} \tilde{r}_{dn\ell} }
\\
&= \sum_{n=1}^{U_d} c_{dn} \frac{\mbox{expElog}[\pi_{dk}] \mbox{expElog}[L_{dnk}]}{\sum_{\ell} \tilde{r}_{dn\ell} }
\\
&= \mbox{expElog}[\pi_{dk}] \sum_{n=1}^{U_d} c_{dn} \mbox{expElog}[L_{dnk}] / \mbox{sum}[ \tilde{r}_{dn}] )
\end{align}$
